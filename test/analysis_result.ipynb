```python
import pickle
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getMin(DistPair):
    Data = DistPair[0]
    return np.min(Data).item()

def getMax(DistPair):
    Data = DistPair[0]
    return np.max(Data).item()

def getStd(DistPair):
    Datas = []
    for val, count in zip(DistPair[0], DistPair[1]):
      Datas += [val] * count
    return np.std(Datas).item()
```
